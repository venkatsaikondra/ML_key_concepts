```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
from sklearn.datasets import make_regression

x, y = make_regression(
    n_samples=4,
    n_features=1,
    n_informative=1,
    n_targets=1,
    noise=80,
    random_state=13
)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red')
```

```python
y_pred=((78.35*x)+0).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred,color='green',label='b=0')
plt.legend()
plt.show()
```

```python
m=78.35
b=0
loss_slope=-2*np.sum(y-m*x.ravel()-b)
loss_slope
```

```python
lr=0.1
step_size=loss_slope*lr
step_size
```

```python
b=b-step_size
b
```

```python
y_pred1=((78.35*x)+b).reshape(4)
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred1,color='orange',label='New b')
plt.plot(x,y_pred,color='green',label='b=0')
plt.legend()
plt.show()
```

```python
loss_slope=-2*np.sum(y-m*x.ravel()-b)
loss_slope
```

```python
step_size=loss_slope*lr
step_size
```

```python
b=b-step_size
```

```python
b
```

```python
y_pred2=((78.35*x)+b).reshape(4)
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred2,color='green',label='b2=25.11')
plt.plot(x,y_pred1,color='orange',label='b1=20.92')
plt.plot(x,y_pred,color='green',label='b=0')
plt.legend()
plt.show()
```
